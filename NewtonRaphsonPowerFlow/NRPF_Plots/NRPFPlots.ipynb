**Imports**

```python
from PowerFlow import PowerData, NRPF
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt
```

**Load case data as given in 5.3.4 of Power Flow Analysis by N.V. Ramana (also found in case3.m)**

```python
case3 = PowerData()
case3.loadcasedata('case3.m')
```

```python
print(case3.buses, '\n')
print(case3.gens, '\n')
print(case3.branches)
```

**Initialize Solvers**

```python
# Classical
pf = NRPF()
# HHL
pf_hhl = NRPF('hhl')
```

**Solve case3 classically**

```python
c3_pf, c3_pf_PAVs = pf.NR(case3, 1e-8)
c3_pf_PAVs_norm = [norm(c3_pf_PAVs[-1]-c3_pf_PAVs[i]) for i in range(c3_pf_PAVs.shape[0])]
plt.plot(c3_pf_PAVs_norm)
plt.show()
```

**Solve case3 classically with NR_PC solver**

```python
c3_pc, c3_pc_PAVs = pf.NR_PC(case3, 1e-8)
c3_pc_PAVs_norm = [norm(c3_pc_PAVs[-1]-c3_pc_PAVs[i]) for i in range(c3_pc_PAVs.shape[0])]
plt.plot(c3_pc_PAVs_norm)
plt.show()
```

**Solve case3 with hhl**

```python
c3_hhl, c3_hhl_PAVs = pf_hhl.NR(case3, 1e-8)
c3_hhl_PAVs_norm = [norm(c3_hhl_PAVs[-1]-c3_hhl_PAVs[i]) for i in range(c3_hhl_PAVs.shape[0])]
plt.plot(c3_hhl_PAVs_norm)
plt.show()
```

**Solve case3 with hhl and NR_PC solver**

```python
c3_pc_hhl, c3_pc_hhl_PAVs = pf_hhl.NR_PC(case3, 1e-8)
c3_pc_hhl_PAVs_norm = [norm(c3_pc_hhl_PAVs[-1]-c3_pc_hhl_PAVs[i]) for i in range(c3_pc_hhl_PAVs.shape[0])]
plt.plot(c3_pc_hhl_PAVs_norm)
plt.show()
```
